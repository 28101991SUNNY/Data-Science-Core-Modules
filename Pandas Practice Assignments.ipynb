#### You're given a record of students with their registration dates and registration IDs. You're also given the marks scored by each student in Physics, Chemistry, and Mathematics. You need to find out the average marks by the students who registered on or after 01/05/2014 i.e. May 1st, 2014. Return a list with the corresponding register IDs of the students sorted by their average marks in all three subjects in increasing order.

    Input Format
    A dataframe with columns ['dates', 'RID', 'Phy', 'Chem', 'Math'] 
    Output Format
    You are required to return a list consisting Reg Ids.

    Sample Input

    For input dataframes you can consider the following code:

```python
import pandas as pd
df = pd.DataFrame({'dates':["2015-12-06", "2011-12-27", "2015-09-07", "2012-12-21", "2020-02-13"],
                   'RID':[498, 721, 375, 464, 813], 
                   'Phy':[22, 45, 1, 65, 22], 
                   'Chem':[52, 56, 32, 50, 24],
                   'Math':[63, 37, 68, 62, 43]}) 
```

```python
df
```

```python
df["dates"] = pd.to_datetime(df["dates"])
```

```python
df
```

```python
# 01/05/2014
```

```python
df[df["dates"]>"2014-05-01"]
```

```python
df["avg_marks"] = df[df["dates"]>"2014-05-01"][["Phy","Chem","Math"]].sum(axis=1)
```

```python
df[df["dates"]>"2014-05-01"].sort_values(by="avg_marks")
```

```python
list(df[df["dates"]>"2014-05-01"].sort_values(by="avg_marks")["RID"])
```

### Who of the following has served as the coach of his/her national men and women team (here teams should necessarily be different for men and women, not duet) in Olympics 2021?

```python
coaches = pd.read_excel('Coaches.xlsx', engine='openpyxl')
```

```python
coaches_sample = coaches.Name.value_counts().to_frame()
coaches_sample = coaches_sample.reset_index()
coaches_sample = coaches_sample.rename(columns={"index": "Name", "Name": "team_count"})
coaches_sample = coaches_sample.merge(coaches, how="left", on="Name")
ans=coaches_sample[coaches_sample["team_count"] > 1]
(ans)
```
