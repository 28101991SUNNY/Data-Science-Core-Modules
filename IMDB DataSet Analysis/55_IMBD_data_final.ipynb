# assess every movie with respect to directors average budget.


```python
# group based transformation
```

```python
#movies :  ADDITIONAL budget than tht directors average budget m
```

```python
def subtract_average(x):
    return x-x.mean()
```

```python
data.groupby("director_name")["budget"].transform(subtract_average)
```

```python
data["assessment"] = data.groupby("director_name")["budget"].transform(subtract_average)
```

```python
data
```

```python
def average_budget(x):
    return x.mean()
```

```python
data.groupby("director_name")["budget"].transform(average_budget)
```

```python
data["average_budget_by_director"]= data.groupby("director_name")["budget"].transform(average_budget)
```

```python
data
```

# for a movie, if it was risky for a director or not!



```python
# budget was even higher than the average revenue of tht particular director
```

```python
def funx(x):
    x["risky"] = (x["budget"] - x["revenue"].mean()) >=  0
    return x
```

```python
risk_ = data.groupby("director_name").apply(funx)
```

```python
risk_.loc[risk_["risky"]==True]
```

```python
data
```

```python
def average_revenue(x):
    return x.mean()
```

```python
data["mean_revenue_of_director"]=data.groupby("director_name")["revenue"].transform(average_revenue)
```

```python
data
```

```python
data["risky"] = data["budget"] > data["mean_revenue_of_director"] 
```

```python
data.loc[data["risky"]==True]
```
